```python
import pandas as pd
data = pd.read_csv("Swarm_Behaviour.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
df = pd.DataFrame()
```

```python
l = ["","Vel","A","S","C"]

for i in range(1,201):
        for j in l:
            df[j+str(i)] = (data["x"+j+str(i)] + data["y"+j+str(i)])/2
        df["nAC"+str(i)] = data["nAC"+str(i)]
        df["nS"+str(i)] = data["nS"+str(i)]
```

```python
df.info()
```

```python
x = df.values
y = data["Swarm_Behaviour"]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
xtr,xt,ytr,yt = train_test_split(x,y,test_size = 0.3,random_state = 1)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()

log_reg.fit(xtr,ytr)
```

```python
out = log_reg.predict(xt)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

confusion_matrix(yt,out)
```

```python
accuracy_score(yt,out)
```

```python
from sklearn.metrics import f1_score

f1_score(yt,out)
```

### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
cs = GaussianNB()

cs.fit(xtr,ytr)
```

```python
out_bayes = cs.predict(xt)
out_bayes
```

```python
cs.predict_log_proba(xt)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score

confusion_matrix(yt,out_bayes)
```

```python
f_bay = f1_score(yt,out_bayes)
a_bay = accuracy_score(yt,out_bayes)
print("f1 score = ",f_bay)
print("acccuracy score = ",a_bay)
```

### K-Nearest Neighbours

```python
from sklearn.neighbors import KNeighborsClassifier
k = [0,0]
for i in range(3,10):
    f = KNeighborsClassifier(i)
    f.fit(xtr,ytr)
    if f.score(xt,yt) > k[1]:
        k[0] = i
        k[1] = f.score(xt,yt)
k
```

```python
f = KNeighborsClassifier(7)
f.fit(xtr,ytr)
out_k = f.predict(xt)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score

confusion_matrix(yt,out_k)
```

```python
f_k = f1_score(yt,out_k)
a_k = accuracy_score(yt,out_k)
print("f1 score = ",f_k)
print("acccuracy score = ",a_k)
```
